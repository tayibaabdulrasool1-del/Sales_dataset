```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("sales.xlsx")
df
```

```python
                                                            # find missing data
df.isnull()
                                                            # sum missing vales
df.isnull().sum()
```

```python
                                                          # fill missing with 0
df=df.fillna(0)
df
```

```python
# which product sale most?

df.groupby("Product")["Quantity"].sum().sort_values(ascending=False)
```

```python
                                                # before conversion it is object
df['Date'] = pd.to_datetime(df['Date'])
                                                # now it become datetime64[ns]
```

```python
                                                # .dt is a datetime accessor in pandas
df['Month'] = df['Date'].dt.month
                                                # .month extracts the month number from each date.
```

```python
df
```

```python
                                                          # drop revenue col
df = df.drop("revenue", axis=1)
df
```

```python
                                            # Sales (Revenue) = Quantity  Price
df["sales"]=df["Quantity"]*df["Price"]
```

```python
# total revenue?
revenue=df["sales"].sum()
revenue
```

```python
# which month has most sales?
df.groupby("Month")["Price"].sum().sort_values(ascending=False)
```

```python
# which product generates most revenue(sales)?

df.groupby("Product")["sales"].sum().sort_values(ascending=False)
```

```python
# Barchat revenue per product?

barchat=sns.barplot(x=df["Product"],y=df["sales"])
plt.xticks(rotation=45)
plt.tight_layout()       # adjust plot to fit labels
plt.show()
barchat
```

```python
# monthly revenue trend?

sns.lineplot(x="Month",y="sales",data=df)
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("sales")
plt.title("Monthly Sales Trend")
plt.tight_layout()
plt.show()
```

```python
# correlation between Quantity,Price,sales?

corr=df[["Quantity","Price","sales"]].corr()
corr
```
